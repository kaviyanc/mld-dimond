# importing libarary

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

# scraping website

```python
url='https://www.flipkart.com/search?q'
```

# website response

```python
response=requests.get(url)
```

```python
response
```

# parsing

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

# headset name

```python
headset=soup.select('.s1Q9rs')
```

```python
headset
```

```python
headset_list=[]
headset_link=[]
```

```python
for item in headset:
    headset_list.append(item.text)
    link='https://www.flipkart.com/search?q'+item["href"]
    headset_list.append(link)
```
